# K Means

```python
# Hopkins Test
```

```python
from sklearn.neighbors import NearestNeighbors
from random import sample
from numpy.random import uniform
import numpy as np
from math import isnan
 
def hopkins(X):
    d = X.shape[1]
    #d = len(vars) # columns
    n = len(X) # rows
    m = int(0.1 * n) 
    nbrs = NearestNeighbors(n_neighbors=1).fit(X.values)
 
    rand_X = sample(range(0, n, 1), m)
 
    ujd = []
    wjd = []
    for j in range(0, m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0),np.amax(X,axis=0),d).reshape(1, -1), 2, return_distance=True)
        ujd.append(u_dist[0][1])
        w_dist, _ = nbrs.kneighbors(X.iloc[rand_X[j]].values.reshape(1, -1), 2, return_distance=True)
        wjd.append(w_dist[0][1])
 
    H = sum(ujd) / (sum(ujd) + sum(wjd))
    if isnan(H):
        print(ujd, wjd)
        H = 0
 
    return H
```

```python
from sklearn.cluster import KMeans
```

```python
# Hopkins Score
hopkins(countrydf)
```

# Silhouette Analysis

```python
from sklearn.metrics import silhouette_score
sse_ = []
for k in range(2, 15):
    kmeans = KMeans(n_clusters=k).fit(countrydf)
    sse_.append([k, silhouette_score(countrydf, kmeans.labels_)])
```

```python
# Silhouette Plot
plt.plot(pd.DataFrame(sse_)[0], pd.DataFrame(sse_)[1]);
```

# 2 has highest point followed by 3

# Elbow Curve

```python
# sum of squared distances
ssd = []
for num_clusters in list(range(1,21)):
    model_clus = KMeans(n_clusters = num_clusters, max_iter=50)
    model_clus.fit(countrydf)
    ssd.append(model_clus.inertia_)

plt.plot(ssd)
```

# 2 to 5 are best values in Elbow Curve

# Comparing both Silhouette plot & Elbow curve, 3 is the best choice for us

# Hierarchical  Clustering

```python
# Libraries imported to perform Hierarchical clustering
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
```

```python
# heirarchical clustering
mergings = linkage(countrydf, method = "single", metric='euclidean')
dendrogram(mergings)
plt.show()
```

```python
mergings = linkage(countrydf, method = "complete", metric='euclidean')
dendrogram(mergings)
plt.show()
```

In the above plot, there are 4 distinguisable clusters. So we take clusters as 4
